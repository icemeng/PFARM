# 4.3.1.Single sample

```python
import matplotlib.pyplot as plt
import numpy.random as npr
```

```python
I=10000
x_norm=npr.normal(loc=0.8,scale=1.6,size=I) #N~(0.8,1.6)
x_logn=npr.lognormal(mean=0.5,sigma=1.0,size=I) #logN~(0.5,1.0)
x_chi=npr.chisquare(df=5,size=I) #Chi~(df=0.5)
x_beta=npr.beta(a=2,b=6,size=I) #B~(a=2,b=6)

plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
plt.hist(x_norm,label='Normal Distribution',bins=20,facecolor='y',edgecolor='k')
plt.xticks(fontsize=13)
plt.xlabel('Sample value',fontsize=13)
plt.yticks(fontsize=13)
plt.ylabel('Frequency',fontsize=13)
plt.legend(loc=0,fontsize=13)
plt.grid()

plt.subplot(2,2,2)
plt.hist(x_logn,label='LogNormal Distribution',bins=20,facecolor='r',edgecolor='k')
plt.xticks(fontsize=13)
plt.xlabel('Sample value',fontsize=13)
plt.yticks(fontsize=13)
plt.ylabel('Frequency',fontsize=13)
plt.legend(loc=0,fontsize=13)
plt.grid()

plt.subplot(2,2,3)
plt.hist(x_chi,label='Chi-squared Distribution',bins=20,facecolor='b',edgecolor='k')
plt.xticks(fontsize=13)
plt.xlabel('Sample value',fontsize=13)
plt.yticks(fontsize=13)
plt.ylabel('Frequency',fontsize=13)
plt.legend(loc=0,fontsize=13)
plt.grid()

plt.subplot(2,2,4)
plt.hist(x_beta,label='Beta Distribution',bins=20,facecolor='c',edgecolor='k')
plt.xticks(fontsize=13)
plt.xlabel('Sample value',fontsize=13)
plt.yticks(fontsize=13)
plt.ylabel('Frequency',fontsize=13)
plt.legend(loc=0,fontsize=13)
plt.grid()

plt.show()
```

# 4.3.2.Mutiple samples

## 1.Stacked

```python
import pandas as pd
import numpy as np

SH_SZ_Index=pd.read_excel('SH_SZ_Index daily(2019-2020).xlsx',engine='openpyxl',sheet_name="Sheet1",header=0,index_col=0)
SH_SZ_Index=np.array(SH_SZ_Index) #Dataframe->List
#print(SH_SZ_Index)
plt.figure(figsize=(9,6))
plt.hist(SH_SZ_Index,label=['SH Index','SZ Index'],stacked=True,edgecolor='k',bins=30)
plt.xticks(fontsize=13)
plt.xlabel('Daily Moving',fontsize=13)
plt.yticks(fontsize=13)
plt.ylabel('Frequency',fontsize=13)
plt.title('Stacked histogram of SH and SZ Index Daily Moving',fontsize=13)
plt.legend(loc=0,fontsize=13)
plt.grid()
plt.show()
```

## 2.Sided

```python
plt.figure(figsize=(9,6))
plt.hist(SH_SZ_Index,label=['SH Index','SZ Index'],edgecolor='k',bins=30) #Default stacked=False
plt.xticks(fontsize=13)
plt.xlabel('Daily Moving',fontsize=13)
plt.yticks(fontsize=13)
plt.ylabel('Frequency',fontsize=13)
plt.title('Sided histogram of SH and SZ Index Daily Moving',fontsize=13)
plt.legend(loc=0,fontsize=13)
plt.grid()
plt.show()
```
