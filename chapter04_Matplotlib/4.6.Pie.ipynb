# IMF SDR Currency Weight

Currency | Weight 
:-----: | :----: 
USD   | 41.73% 
EUR   | 30.93%
CNY   | 10.92%
JPY   | 8.33%
GBP   | 8.09%


```python
import matplotlib.pyplot as plt

currency=['USD','EUR','CNY','JPY','GBP']
perc=[0.4173,0.3093,0.1092,0.0833,0.0809]

plt.figure(figsize=(9,7))
plt.pie(x=perc,labels=currency,textprops={'fontsize':13})
plt.axis('equal') #Let the pie chart be circle
plt.title('SDR Currency Weight',fontsize=13)
plt.legend(loc=2,fontsize=13)
plt.show()
```
