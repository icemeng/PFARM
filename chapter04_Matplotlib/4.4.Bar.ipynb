```python
import numpy as np
import pandas as pd

R_array=np.array([[-0.035099,0.017230,-0.003450,-0.024551,0.039368],
                        [-0.013892,0.024334,-0.033758,0.014622,0.000128],
                        [0.005848,-0.002907,0.005831,0.005797,-0.005764],
                        [0.021242,0.002133,-0.029803,-0.002743,-0.014301]])
date=['2020-05-25','2020-05-26','2020-05-27','2020-05-28','2020-05-29']
name=['CAST','CSS','BOC','SAIC']
R_dataframe=pd.DataFrame(data=R_array.T,index=date,columns=name)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
plt.bar(x=R_dataframe.columns,height=R_dataframe.iloc[0],width=0.5,label='2020/05/25',facecolor='y')
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.ylabel('Return',fontsize=13)
plt.legend(loc=0,fontsize=13)
plt.grid()

plt.subplot(2,2,2,sharex=plt.subplot(2,2,1),sharey=plt.subplot(2,2,1))
plt.bar(x=R_dataframe.columns,height=R_dataframe.iloc[1],width=0.5,label='2020/05/26',facecolor='c')
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(loc=0,fontsize=13)
plt.grid()

plt.subplot(2,2,3,sharex=plt.subplot(2,2,1),sharey=plt.subplot(2,2,1))
plt.bar(x=R_dataframe.columns,height=R_dataframe.iloc[3],width=0.5,label='2020/05/28',facecolor='b')
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.ylabel('Return',fontsize=13)
plt.legend(loc=0,fontsize=13)
plt.grid()

plt.subplot(2,2,4,sharex=plt.subplot(2,2,1),sharey=plt.subplot(2,2,1))
plt.bar(x=R_dataframe.columns,height=R_dataframe.iloc[4],width=0.5,label='2020/05/29',facecolor='g')
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(loc=0,fontsize=13)
plt.grid()
plt.show()
```

# 4.4.2.Horizontal

```python
plt.figure(figsize=(9,6))

plt.barh(y=R_dataframe.columns,width=R_dataframe.iloc[1],height=0.5,label='2020/05/26')
plt.barh(y=R_dataframe.columns,width=R_dataframe.iloc[2],height=0.5,label='2020/05/27')
plt.xticks(fontsize=13)
plt.xlabel('Return',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Horizontal bar chart of 4 stocks',fontsize=13)
plt.legend(loc=1,fontsize=13)
plt.grid()
plt.show()
```

# 4.4.3.Biaxial

```python
M2=pd.read_excel('M2(2019-2020).xlsx',engine='openpyxl',sheet_name="Sheet1",header=0,index_col=0)
```

```python
fig,ax1=plt.subplots(figsize=(9,6))
plt.bar(x=M2.index,height=M2.iloc[:,0],color='y',label='M2 Month end balance')
plt.xticks(fontsize=13,rotation=90)
plt.xlabel('Date',fontsize=13)
plt.yticks(fontsize=13)
plt.ylim(0,250)
plt.ylabel('Amount(Trillion)',fontsize=13)
plt.legend(loc=2,fontsize=13)

ax2=ax1.twinx()
plt.plot(M2.iloc[:,-1],label='M2 Monthly growth rate',lw=2.5)
plt.yticks(fontsize=13)
plt.ylim(0,0.13)
plt.ylabel('Growth rate',fontsize=13)
plt.title('M2 Month End Balance and Growth Rate',fontsize=13)
plt.legend(loc=1,fontsize=13)
plt.grid()
plt.show()
```
