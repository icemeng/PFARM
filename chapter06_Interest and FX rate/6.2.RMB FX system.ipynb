# 6.2.2.RMB FX product

```python
import pandas as pd
import matplotlib.pyplot as plt

exchange=pd.read_excel('RMB FX daily(20050721-2020).xlsx',engine='openpyxl',sheet_name="Sheet1",header=0,index_col=0)
exchange.plot(subplots=True,sharex=True,layout=(2,2),figsize=(11,9),grid=True,fontsize=13)
plt.subplot(2,2,1)
plt.ylabel('FX rate',fontsize=13,position=(0,0))
```
