# 9.1.1.IRS market

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

IRS_data=pd.read_excel('IRS market.xlsx',engine='openpyxl',sheet_name="Sheet1",header=0,index_col=0)

name=IRS_data.index
volume=(np.array(IRS_data)).ravel() #transform to 1-D array

plt.figure(figsize=(9,7))
plt.pie(x=volume,labels=name,textprops={'fontsize':13})
plt.axis('equal')
plt.show()
```
