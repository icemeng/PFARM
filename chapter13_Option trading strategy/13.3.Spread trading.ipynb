# 13.3.1.Bull spread

**Use call option**\
Long strike K1 call option + Short strike K2 call option(K1<K2)

```python
import numpy as np
import matplotlib.pyplot as plt

K1=4500
K2=5000
C1=474.4
C2=293.0

S0=4762.76
St=np.linspace(3500,6500,500)

N_1=1
N_2=1
M=100

profit_C1_long=N_1*M*(np.maximum(St-K1,0)-C1)
profit_C2_short=N_2*M*(C2-np.maximum(St-K2,0))
profit_bullspread=profit_C1_long+profit_C2_short

plt.figure(figsize=(9,6))
plt.plot(St,profit_C1_long,'b--',label='Long strike K1 call option',lw=2.5)
plt.plot(St,profit_C2_short,'g--',label='Short strike K2 call option',lw=2.5)
plt.plot(St,profit_bullspread,'r-',label='Bull spread strategy',lw=2.5)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and bull spread strategy',fontsize=13)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```

**Use put option**\
Long strike K1 put option + Short strike K2 put option(K1<K2)

```python
K1=4400
K2=5200
P1=290.0
P2=873.8

S0=4812.76
St=np.linspace(3000,6000,500)

profit_P1_long=N_1*M*(np.maximum(K1-St,0)-P1)
profit_P2_short=N_2*M*(P2-np.maximum(K2-St,0))
profit_bullspread=profit_P1_long+profit_P2_short

plt.figure(figsize=(9,6))
plt.plot(St,profit_P1_long,'b--',label='Long strike K1 put option',lw=2.5)
plt.plot(St,profit_P2_short,'g--',label='Short strike K2 put option',lw=2.5)
plt.plot(St,profit_bullspread,'r-',label='Bull spread strategy',lw=2.5)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and bull spread strategy',fontsize=13)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```

# 13.3.3.Box spread

Bull spread using call option - Bear spread using put option

```python
K1=4000
K2=4600
C1=161.6
C2=33.2
P1=285.4
P2=776.0

S0=4044.38
St=np.linspace(3000,5000,500)

profit_C1_long=N_1*M*(np.maximum(St-K1,0)-C1)
profit_C2_short=N_2*M*(C2-np.maximum(St-K2,0))
profit_P1_short=N_1*M*(P1-np.maximum(K1-St,0))
profit_P2_long=N_2*M*(np.maximum(K2-St,0)-P2)

profit_boxspread=profit_C1_long+profit_C2_short+profit_P1_short+profit_P2_long

plt.figure(figsize=(9,6))
plt.plot(St,profit_C1_long,'b--',label='Long strike K1 call option',lw=2.5)
plt.plot(St,profit_C2_short,'g--',label='Short strike K2 call option',lw=2.5)
plt.plot(St,profit_P1_short,'c--',label='Short strike K1 put option',lw=2.5)
plt.plot(St,profit_P2_long,'m--',label='Long strike K2 put option',lw=2.5)

plt.plot(St,profit_boxspread,'r-',label='Box spread strategy',lw=2.5)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and box spread strategy',fontsize=13)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```

```python
shibor=0.02115
tenor=0.5

PV_boxspread=profit_boxspread[0]*np.exp(-shibor*tenor)
print('PV of box spread is ',round(PV_boxspread,2))
```

# 13.3.4.Butterfly spread

**Use call option**\
Long strike K1 call option + Short 2 strike K2 call option + Long strike K3 call option(K2=(K1+K3)/2)

```python
K1=4400
K2=4800
K3=5200
C1=571.6
C2=388.6
C3=255.0

S0=4842.12
St=np.linspace(3400,6200,500)

N1=1
N2=2
N3=1

profit_C1_long=N1*M*(np.maximum(St-K1,0)-C1)
profit_C2_short=N2*M*(C2-np.maximum(St-K2,0))
profit_C3_long=N3*M*(np.maximum(St-K3,0)-C3)
profit_buttspread=profit_C1_long+profit_C2_short+profit_C3_long

plt.figure(figsize=(9,6))
plt.plot(St,profit_C1_long,'b--',label='Long strike K1 call option',lw=2.5)
plt.plot(St,profit_C2_short,'g--',label='Short strike K2 call option',lw=2.5)
plt.plot(St,profit_C3_long,'c--',label='Long strike K3 call option',lw=2.5)
plt.plot(St,profit_buttspread,'r-',label='Butterfly spread strategy',lw=2.5)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and Butterfly spread strategy',fontsize=13)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```

**Use put option**\
Long strike K1 put option + Short 2 strike K2 put option + Long strike K3 put option(K2=(K1+K3)/2)

```python
K1=4200
K2=4600
K3=5000
P1=264.2
P2=476.2
P3=748.2

S0=4581.98
St=np.linspace(3200,6000,500)

profit_P1_long=N1*M*(np.maximum(K1-St,0)-P1)
profit_P2_short=N2*M*(P2-np.maximum(K2-St,0))
profit_P3_long=N3*M*(np.maximum(K3-St,0)-P3)
profit_buttspread=profit_P1_long+profit_P2_short+profit_P3_long

plt.figure(figsize=(9,6))
plt.plot(St,profit_P1_long,'b--',label='Long strike K1 put option',lw=2.5)
plt.plot(St,profit_P2_short,'g--',label='Short strike K2 put option',lw=2.5)
plt.plot(St,profit_P3_long,'c--',label='Long strike K3 put option',lw=2.5)
plt.plot(St,profit_buttspread,'r-',label='Butterfly spread strategy',lw=2.5)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and Butterfly spread strategy',fontsize=13)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```

# 13.4.5.Calendar spread

**Use call option**\
Short near maturity call option + Long longer maturity call option

```python
import utils

K_same=4600
C1=224.0
C2=348.0

S0=4584.59
St=np.linspace(3000,6000,500)

N1=1
N2=1

profit_C1_short=N1*M*(C1-np.maximum(St-K_same,0))

tenor=0.5
sigma_index=0.22
shibor=0.02922
step=120

Ct=utils.BTM_Nstep(S=St,K=K_same,sigma=sigma_index,r=shibor,T=tenor,N=step,types='call')

profit_C2_long=N2*M*(Ct-C2)
profit_calendarspread=profit_C1_short+profit_C2_long

plt.figure(figsize=(9,6))
plt.plot(St,profit_C1_short,'b--',label='Short near maturity call option',lw=2.5)
plt.plot(St,profit_C2_long,'g--',label='Long longer maturity call option',lw=2.5)
plt.plot(St,profit_calendarspread,'r-',label='Calendar spread strategy',lw=2.5)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and calendar spread strategy',fontsize=13)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```

**Use put option**\
Short near maturity put option + Long longer maturity put option

```python
K_same=5000
P1=597.6
P2=746.4
St=np.linspace(3500,6500,500)

profit_P1_short=N1*M*(P1-np.maximum(K_same-St,0))

Pt=utils.BTM_Nstep(S=St,K=K_same,sigma=sigma_index,r=shibor,T=tenor,N=step,types='put')

profit_P2_long=N2*M*(Pt-P2)
profit_calendarspread=profit_P1_short+profit_P2_long

plt.figure(figsize=(9,6))
plt.plot(St,profit_P1_short,'b--',label='Short near maturity put option',lw=2.5)
plt.plot(St,profit_P2_long,'g--',label='Long longer maturity put option',lw=2.5)
plt.plot(St,profit_calendarspread,'r-',label='Calendar spread strategy',lw=2.5)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and calendar spread strategy',fontsize=13)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```
