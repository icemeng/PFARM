# 13.2.1.Long covered call

long covered call=short underlying asset + long call opion

```python
import numpy as np
import matplotlib.pyplot as plt

C=0.2065
K=4.0
S0_ETF=4.0364
S0_index=3971.34

St_index=np.linspace(3000,5000,500)
St_ETF=S0_ETF*St_index/S0_index

N_ETF=10000
N_call=1
N_underlying=10000

profit_ETF_short=-N_ETF*(St_ETF-S0_ETF)
profit_call_long=N_call*N_underlying*(np.maximum(St_ETF-K,0)-C)
profit_covcall_long=profit_ETF_short+profit_call_long

plt.figure(figsize=(9,6))
plt.plot(St_index,profit_ETF_short,'b--',label='Short HS300 ETF',lw=2.5)
plt.plot(St_index,profit_call_long,'g--',label='Long HS300 ETF call option',lw=2.5)
plt.plot(St_index,profit_covcall_long,'r-',label='Long covered call',lw=2.5)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and long covered call',fontsize=13)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```

# 13.2.2.Short covered call

short(writing) covered call=long underlying asset + short call opion

```python
C=0.201
K=4.3
S0_ETF=4.3429
S0_index=4247.78

St_index=np.linspace(3500,5500,500)
St_ETF=S0_ETF*St_index/S0_index

profit_ETF_long=N_ETF*(St_ETF-S0_ETF)
profit_call_short=-N_call*N_underlying*(np.maximum(St_ETF-K,0)-C)
profit_covcall_short=profit_ETF_long+profit_call_short

plt.figure(figsize=(9,6))
plt.plot(St_index,profit_ETF_long,'b--',label='Long HS300 ETF',lw=2.5)
plt.plot(St_index,profit_call_short,'g--',label='Short HS300 ETF call option',lw=2.5)
plt.plot(St_index,profit_covcall_short,'r-',label='Short covered call',lw=2.5)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and short covered call',fontsize=13)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```

# 13.2.3.Long protective put

long protective put=long underlying asset + long put opion

```python
P=0.4416
K=4.9
S0_ETF=4.9168
S0_index=4771.31

St_index=np.linspace(4000,6000,500)
St_ETF=S0_ETF*St_index/S0_index

N_put=1

profit_ETF_long=N_ETF*(St_ETF-S0_ETF)
profit_put_long=N_put*N_underlying*(np.maximum(K-St_ETF,0)-P)
profit_protput_long=profit_ETF_long+profit_put_long

plt.figure(figsize=(9,6))
plt.plot(St_index,profit_ETF_long,'b--',label='Long HS300 ETF',lw=2.5)
plt.plot(St_index,profit_put_long,'g--',label='Long HS300 ETF put option',lw=2.5)
plt.plot(St_index,profit_protput_long,'r-',label='Long protective put',lw=2.5)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and long protective put',fontsize=13)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```

# 13.2.4.Short protective put

short protective put=short underlying asset + short put opion

```python
P=0.4211
K=5.0
S0_ETF=4.9966
S0_index=4842.12

St_index=np.linspace(3800,5800,500)
St_ETF=S0_ETF*St_index/S0_index

N_put=1

profit_ETF_short=-N_ETF*(St_ETF-S0_ETF)
profit_put_short=-N_put*N_underlying*(np.maximum(K-St_ETF,0)-P)
profit_protput_short=profit_ETF_short+profit_put_short

plt.figure(figsize=(9,6))
plt.plot(St_index,profit_ETF_short,'b--',label='Short HS300 ETF',lw=2.5)
plt.plot(St_index,profit_put_short,'g--',label='Short HS300 ETF put option',lw=2.5)
plt.plot(St_index,profit_protput_short,'r-',label='Short protective put',lw=2.5)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and short protective put',fontsize=13)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```
