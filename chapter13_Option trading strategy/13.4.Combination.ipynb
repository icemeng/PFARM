# 13.4.1.Straddle

**Bottom straddle(straddle purchase)**\
Long call option + Long put option

```python
import numpy as np
import matplotlib.pyplot as plt

K=4700
C=336.4
P=326.4

S0=4698.13
St=np.linspace(3000,6400,500)

N_C=1
N_P=1
M=100

profit_C_long=N_C*M*(np.maximum(St-K,0)-C)
profit_P_long=N_P*M*(np.maximum(K-St,0)-P)
profit_straddle=profit_C_long+profit_P_long

plt.figure(figsize=(9,6))
plt.plot(St,profit_C_long,'b--',label='Long call option',lw=2.5)
plt.plot(St,profit_P_long,'g--',label='Long put option',lw=2.5)
plt.plot(St,profit_straddle,'r-',label='Bottom straddle strategy',lw=2.5)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and bottom straddle strategy',fontsize=13)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```

**Top straddle(Writing straddle)**\
Short call option + Short put option

```python
plt.figure(figsize=(9,6))
plt.plot(St,-profit_C_long,'b--',label='Short call option',lw=2.5)
plt.plot(St,-profit_P_long,'g--',label='Short put option',lw=2.5)
plt.plot(St,-profit_straddle,'r-',label='Top straddle strategy',lw=2.5)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and top straddle strategy',fontsize=13)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```

# 13.4.2.Strip and strap

**Strip**\
Long call option + Long 2 put option

**Strap**\
Long 2 call option + Long put option

```python
K=4600
C=290.0
P=387.0

S0=4635.71
St=np.linspace(3000,6200,500)

N1=1
N2=2

profit_C_strip=N1*M*(np.maximum(St-K,0)-C)
profit_P_strip=N2*M*(np.maximum(K-St,0)-P)
profit_strip=profit_C_strip+profit_P_strip

profit_C_strap=N2*M*(np.maximum(St-K,0)-C)
profit_P_strap=N1*M*(np.maximum(K-St,0)-P)
profit_strap=profit_C_strap+profit_P_strap

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(St,profit_C_strip,'b--',label='Long call option',lw=2.0)
plt.plot(St,profit_P_strip,'c--',label='Long 2 put option',lw=2.0)
plt.plot(St,profit_strip,'r-',label='Strip strategy',lw=2.0)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and strip strategy',fontsize=13)
plt.legend(loc=0,fontsize=13)
plt.grid()

plt.subplot(1,2,2)
plt.plot(St,profit_C_strap,'b--',label='Long 2 call option',lw=2.0)
plt.plot(St,profit_P_strap,'c--',label='Long put option',lw=2.0)
plt.plot(St,profit_strap,'r-',label='Strap strategy',lw=2.0)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and strap strategy',fontsize=13)
plt.legend(loc=0,fontsize=13)
plt.grid()
plt.show()
```

# 13.4.3.Strangle

**Long strangle(bottom vertical combination)**\
Long strike K2 call option + Long strike K1 put option(K1 < K2)

```python
K1=4200
K2=4900
C=245.0
P=264.2

S0=4581.98
St=np.linspace(3000,6100,500)

N_C=1
N_P=1

profit_C_long=N_C*M*(np.maximum(St-K2,0)-C)
profit_P_long=N_P*M*(np.maximum(K1-St,0)-P)
profit_strangle=profit_C_long+profit_P_long

plt.figure(figsize=(9,6))
plt.plot(St,profit_C_long,'b--',label='Long call option',lw=2.5)
plt.plot(St,profit_P_long,'g--',label='Long put option',lw=2.5)
plt.plot(St,profit_strangle,'r-',label='Long strangle strategy',lw=2.5)
plt.xlabel('HS300 index',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between HS300 index and long strangle strategy',fontsize=13)
plt.legend(loc=9,fontsize=13)
plt.grid()
plt.show()
```

**Short strangle(top vertical combination)**\
Short strike K2 call option + Short strike K1 put option(K1 < K2)

```python
K1=1.9
K2=2.0
C=0.0104
P=0.0116

V1=K1-P-C #breakeven point 1
V2=K2+P+C #breakeven point 2
print('Short strangle breakeven point1 is ',round(V1,4))
print('Short strangle breakeven point2 is ',round(V2,4))
```

```python
St=np.linspace(1.6,2.3,100)

profit_C_short=N_C*(C-np.maximum(St-K2,0))
profit_P_short=N_P*(P-np.maximum(K1-St,0))
profit_strangle_short=profit_C_short+profit_P_short

plt.figure(figsize=(9,6))
plt.plot(St,profit_C_short,'b--',label='Short call option',lw=2.0)
plt.plot(St,profit_P_short,'g--',label='Short put option',lw=2.0)
plt.plot(St,profit_strangle_short,'r-',label='Short strangle strategy',lw=2.5)
plt.xlabel('FX rate(USD/GBP)',fontsize=13)
plt.xticks(fontsize=13)
plt.ylabel('Profit',fontsize=13)
plt.yticks(fontsize=13)
plt.title('Relationship between FX rate and short strangle strategy',fontsize=13)
plt.annotate('Breakeven point1',xy=(V1,0.0),xytext=(1.58,-0.01),arrowprops=dict(shrink=0.01),fontsize=13)
plt.annotate('Breakeven point2',xy=(V2,0.0),xytext=(2.08,-0.01),arrowprops=dict(shrink=0.01),fontsize=13)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```
