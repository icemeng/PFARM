# 15.1.2.VaR Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

x=0.95 #95% Confidence level
z=st.norm.ppf(q=1-x) #Percent Point Function
x=np.linspace(-4,4,200)
y=st.norm.pdf(x) #Probability Density Function
x1=np.linspace(-4,z,100)
y1=st.norm.pdf(x1)

plt.figure(figsize=(9,6))
plt.plot(x,y,'r-',lw=2.0)
plt.fill_between(x1,y1) #Plot dash area
plt.xlabel('Profit/Loss',fontsize=13)
plt.ylabel('Probability Density',fontsize=13)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.ylim(0,0.42)
plt.annotate('VaR', xy=(z-0.02, st.norm.pdf(z)+0.005),xytext=(-2.3,0.17),arrowprops=dict(shrink=0.01),fontsize=13)
plt.title('VaR(Normal Distribution)',fontsize=13)
plt.grid('True')
plt.show()
```
