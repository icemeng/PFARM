```python
import pandas as pd
import matplotlib.pyplot as plt

bond_GDP=pd.read_excel('Bond and GDP(2010-2020).xlsx',engine='openpyxl',sheet_name="Sheet1",header=0,index_col=0)
bond_GDP.plot(kind='bar',figsize=(9,6),fontsize=13,grid=True)
plt.ylabel('Amount', fontsize=11)
```

```python
bond=pd.read_excel('Bond market distribution 2020.xlsx',engine='openpyxl',sheet_name="Sheet1",header=0,index_col=0)
plt.figure(figsize=(15,6))
plt.pie(x=bond['Bond storage(trillion)'],labels=bond.index)
plt.axis('equal')
plt.legend(loc=2,fontsize=13)
plt.title('Bond market distribution 2020 end',fontsize=13)
plt.show()
```
