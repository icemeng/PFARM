# 10.1.2.Commodity futures

```python
import pandas as pd
import matplotlib.pyplot as plt

data_AU2008=pd.read_excel('AU2008.xlsx',engine='openpyxl',sheet_name="Sheet1",header=0,index_col=0)

data_AU2008.plot(figsize=(10,9),subplots=True,layout=(2,2),grid=True,fontsize=13)
plt.subplot(2,2,1)
plt.ylabel('Price or Amount',fontsize=11,position=(0,0))
```
